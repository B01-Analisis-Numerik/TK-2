```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load CSV file
x_vandermonde     = pd.read_csv('Evaluasi/vandermonde_x.csv')['x(t)']
x_newton          = pd.read_csv('Evaluasi/newton_x.csv')['x(t)']
x_cubic_spline    = pd.read_csv('Evaluasi/cubic_spline_x.csv')['x(t)']

y_vandermonde     = pd.read_csv('Evaluasi/vandermonde_y.csv')['y(t)']
y_newton          = pd.read_csv('Evaluasi/newton_y.csv')['y(t)']
y_cubic_spline    = pd.read_csv('Evaluasi/cubic_spline_y.csv')['y(t)']

t_eval = np.linspace(0, 540, 541)
```

```python
# Plot Ketiga Metode dalam Satu Grafik

plt.figure(figsize=(14, 7))

# Subplot Untuk Interpolasi x
plt.subplot(1, 2, 1)
plt.plot(t_eval, x_vandermonde, '-', label='Vandermonde', color='blue')
plt.plot(t_eval, x_newton, '--', label='Newton', color='red')
plt.plot(t_eval, x_cubic_spline, ':', label='Cubic Spline', color='green')
plt.title('Interpolasi x')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()

# Subplot Untuk Interpolasi y
plt.subplot(1, 2, 2)
plt.plot(t_eval, y_vandermonde, '-', label='Vandermonde', color='blue')
plt.plot(t_eval, y_newton, '--', label='Newton', color='red')
plt.plot(t_eval, y_cubic_spline, ':', label='Cubic Spline', color='green')
plt.title('Interpolasi y')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()

plt.show()
```
